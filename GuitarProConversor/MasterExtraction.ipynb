```python
import os
import glob
import guitarpro
import re

# Function to find all GuitarPro files recursively
def find_gp_files(folder_path):
    gp_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".gp") or file.endswith(".gp2") or file.endswith(".gp3") or file.endswith(".gp4") or file.endswith(".gp5") and file.split('.') == 1:
                gp_files.append(os.path.join(root, file))
    return gp_files


def remove_problematic_symbols(text):
    # Define problematic symbols or characters you want to remove
    problematic_symbols = r'[!@#$%^&*()_+={}\[\]:;"\'|\\<,>.?/~`]'
    
    # Remove problematic symbols
    cleaned_text = re.sub(problematic_symbols, '', text)
    
    return cleaned_text

def compact_string(text):
    # Remove leading and trailing whitespaces
    text = text.strip()
    
    # Replace multiple whitespaces with a single whitespace
    text = re.sub(r'\s+', ' ', text)
    
    return text


main_folder_path = 'DummyTabs/'

# Get all GuitarPro files recursively
gp_files = find_gp_files(main_folder_path)

OutputFolder = 'tabs/'

# print(len(gp_files))

# Iterate and export all guitar tracks in each GuitarPro file
counter = 1 
for file_path in gp_files:
    # Open the file
    gp_file = guitarpro.parse(file_path)
    
    # Extract guitar tracks
    tracks = [track for track in gp_file.tracks]
    
    guitar_tracks = []
    for track in tracks:
        if  track.channel.instrument >= 24 and track.channel.instrument <= 31:
            guitar_tracks.append(track)
            # print(track.name, track)
    
    for track in guitar_tracks:
        file_extension = '.gp3' 
        
        track_gp_file = guitarpro.Song(tracks=[track])
        track_file_path = OutputFolder + str(counter) + '.gp3'
        # if os.path.exists(track_file_path):
        #     track_file_path = f'{OutputFolder + compact_string(remove_problematic_symbols(track.name))}_{counter}{file_extension}'
        #     counter += 1
        #     export_file = guitarpro.write(song=track_gp_file, stream=track_file_path)
        # else:
        #     track_file_path = f'{OutputFolder + compact_string(remove_problematic_symbols(track.name))}{file_extension}'
        #     export_file = guitarpro.write(song=track_gp_file, stream=track_file_path)
        export_file = guitarpro.write(song=track_gp_file, stream=track_file_path)
        counter += 1
    # print(track_gp_file)
```
